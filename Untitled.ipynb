# CLUSTERING-univariate,bivariate,multivariate

```python
c1 = KMeans(n_clusters=3)
```

```python
c1.fit(df[['Annual Income ($)']])
```

```python
c1.labels_
```

```python
df['Income Cluster'] = c1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
c1.inertia_
```

```python
intertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income ($)']])
    intertia_scores.append(kmeans.inertia_)
print(intertia_scores)
```

```python
plt.plot(range(1,11),intertia_scores)
```

```python
df.columns
```

```python
df.groupby('Income Cluster')[['Age', 'Annual Income ($)','Spending Score (1-100)']].mean()
```

```python
#Bivariate Clustering
c2 = KMeans(n_clusters=5)
c2.fit(df[['Annual Income ($)','Spending Score (1-100)']])
df['Spending and Income Cluster'] =c2.labels_
df.head()
```

```python
intertia_scores2=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(df[['Annual Income ($)','Spending Score (1-100)']])
    intertia_scores2.append(kmeans2.inertia_)
plt.plot(range(1,11),intertia_scores2)
```

```python
centers =pd.DataFrame(c2.cluster_centers_)
centers.columns = ['x','y']
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c='black',marker='*')
sns.scatterplot(data=df, x ='Annual Income ($)',y='Spending Score (1-100)',hue='Spending and Income Cluster',palette='tab10')
plt.savefig('clustering_bivaraiate.png')
```

```python
pd.crosstab(df['Spending and Income Cluster'],df['Gender'],normalize='index')
```

```python
df.groupby('Spending and Income Cluster')[['Age', 'Annual Income ($)','Spending Score (1-100)']].mean()
```

```python
#mulivariate clustering 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
df.head()
```

```python
dff = pd.get_dummies(df,drop_first=True)
dff.head()
```

```python
dff.columns
```

```python
dff = dff[['Age', 'Annual Income ($)', 'Spending Score (1-100)','Gender_Male']]
dff.head()
```

```python
dff = scale.fit_transform(dff)
dff = pd.DataFrame(scale.fit_transform(dff))
dff.head()
```

```python
intertia_scores3=[]
for i in range(1,11):
    kmeans3=KMeans(n_clusters=i)
    kmeans3.fit(dff)
    intertia_scores3.append(kmeans3.inertia_)
plt.plot(range(1,11),intertia_scores3)
```
